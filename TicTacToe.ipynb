```python
# Tic Tac Toe
 
import random
 
def tablero(board):
   #esta función simula la interfaz de un tablero de triki
    print('    ┃    ┃')
    print(' ' + board[7] + '  ┃ ' + board[8] + '  ┃ ' + board[9])
    print('    ┃    ┃')
    print('――――――――――――――――')
    print('    ┃    ┃')
    print(' ' + board[4] + '  ┃ ' + board[5] + '  ┃ ' + board[6])
    print('    ┃    ┃')
    print('――――――――――――――――')
    print('    ┃    ┃')
    print(' ' + board[1] + '  ┃ ' + board[2] + '  ┃ ' + board[3])
    print('    ┃    ┃')
 
def entrada_usuario():
    
    letra = ''
    while not (letra == 'X' or letra == 'O'):
        print('¿Quieres ser X or O?')
        letra = input().upper()
 
   
    if letra == 'X':
        return ['X', 'O']
    else:
        return ['O', 'X']
 
def primero():
   #escoge al azar quien va primero
    if random.randint(0, 1) == 0:
        return 'IA'
    else:
        return 'Player'
 
def otravez():
  
    print('Quieres jugar de nuevo? (si o no)')
    return input().lower().startswith('s')

def makeMove(board, letra, move):
    board[move] = letra
 
def isWinner(bo, le):
    #posibilidades
    return ((bo[7] == le and bo[8] == le and bo[9] == le) or 
    (bo[4] == le and bo[5] == le and bo[6] == le) or 
    (bo[1] == le and bo[2] == le and bo[3] == le) or 
    (bo[7] == le and bo[4] == le and bo[1] == le) or 
    (bo[8] == le and bo[5] == le and bo[2] == le) or 
    (bo[9] == le and bo[6] == le and bo[3] == le) or 
    (bo[7] == le and bo[5] == le and bo[3] == le) or 
    (bo[9] == le and bo[5] == le and bo[1] == le)) 

def getBoardCopy(board):

    dupeBoard = []
 
    for i in board:
        dupeBoard.append(i)
 
    return dupeBoard
 
def isSpaceFree(board, move):

    return board[move] == ' '
 
def getPlayerMove(board):
  
    move = ' '
    while move not in '1 2 3 4 5 6 7 8 9'.split() or not isSpaceFree(board, int(move)):
        print('Elige una casilla (1-9)')
        move = input()
    return int(move)
 
def chooseRandomMoveFromList(board, movesList):
  
    possibleMoves = []
    for i in movesList:
        if isSpaceFree(board, i):
            possibleMoves.append(i)
 
    if len(possibleMoves) != 0:
        return random.choice(possibleMoves)
    else:
        return None
 
def getComputerMove(board, computerLetter):

    if computerLetter == 'X':
        playerLetter = 'O'
    else:
        playerLetter = 'X'
 #IA
    for i in range(1, 10):
        copy = getBoardCopy(board)
        if isSpaceFree(copy, i):
            makeMove(copy, computerLetter, i)
            if isWinner(copy, computerLetter):
                return i
 
    for i in range(1, 10):
        copy = getBoardCopy(board)
        if isSpaceFree(copy, i):
            makeMove(copy, playerLetter, i)
            if isWinner(copy, playerLetter):
                return i
 

    move = chooseRandomMoveFromList(board, [1, 3, 7, 9])
    if move != None:
        return move
 

    if isSpaceFree(board, 5):
        return 5
 
 
    return chooseRandomMoveFromList(board, [2, 4, 6, 8])
 
def isBoardFull(board):
   
    for i in range(1, 10):
        if isSpaceFree(board, i):
            return False
    return True


print('Ticcccccctaccccccctoeeeee')
 
while True:

    theBoard = [' '] * 10
    playerLetter, computerLetter = entrada_usuario()
    turn = primero()
    print('El ' + turn + ' va primero')
    gameIsPlaying = True
 
    while gameIsPlaying:
        if turn == 'player':
        
            tablero(theBoard)
            move = getPlayerMove(theBoard)
            makeMove(theBoard, playerLetter, move)
 
            if isWinner(theBoard, playerLetter):
                tablero(theBoard)
                print('Ganaste!')
                gameIsPlaying = False
            else:
                if isBoardFull(theBoard):
                    tablero(theBoard)
                    print('Empate')
                    break
                else:
                    turn = 'computer'

        
        
        else:
            # Computer's turn.
            move = getComputerMove(theBoard, computerLetter)
            makeMove(theBoard, computerLetter, move)
 
            if isWinner(theBoard, computerLetter):
                tablero(theBoard)
                print('Perdiste')
                gameIsPlaying = False
            else:
                if isBoardFull(theBoard):
                    tablero(theBoard)
                    print('Empate')
                    break
                else:
                    turn = 'player'
 
    if not otravez():
        break
 
```
